# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python

plt.figure(figsize=(5,5))
plt.hist(cars_data['Buyer Gender'],rwidth=0.8 )
plt.title('Participants Genderwise') 
plt.xticks()
plt.yticks()
plt.grid(True)
```

# 4. Find the top 5 cars based on their sales price.

```python
df=pd.DataFrame(cars_data)  
```

```python
cars_data1= df.groupby('Model').sum('Sale Price').sort_values('Sale Price',ascending=False)
cars_data1.head(5)
```

# 5. Find the least 5 cars based on their Resell price.

```python
cars_data3= df.groupby(by='Model').sum('Resell Price').sort_values('Resell Price',ascending=True)
```
